**Errores en los Tipos de Datos**

```python
df.dtypes
```

sale_price, land_square_feet y gross_square_feet almacenan datos de tipo texto, lo cual no hace mucho sentido porque su nombre indica que deberían ser de punto flotante.

```python
def convertir_flotante(columna):
    
    return pd.to_numeric(columna, errors="coerce")

df["sale_price"] = convertir_flotante(df["sale_price"])
df["land_square_feet"] = convertir_flotante(df["land_square_feet"])
df["gross_square_feet"] = convertir_flotante(df["gross_square_feet"])

df.dtypes
```

**Diccionario de Datos del DataFrame Preprocesado**

```python
df_sucio = pd.read_csv("miaad-nyc-r-s.csv", engine="python", skipfooter=1)

diccionario_datos = pd.DataFrame({"Nombre original de columna" : df_sucio.columns,
                                  "Nombre nuevo de columna" : df.columns,
                                   "Tipo original de dato" : [str(df_sucio[col].dtype) for col in df_sucio.columns],
                                   "Nuevo tipo de dato" : [str(df[col].dtype) for col in df.columns]})
diccionario_datos
```
